### Part B. Characters in Hamlet 
Use the text available at https://s3.amazonaws.com/dmfa-2020/project-1/hamlet.txt for this part.
How many times do each of the following characters have speaking lines in Hamlet? Keep in
mind that this is the text of a play.
Hamlet, Polonius, Ophelia, and Horatio

```python
!wget https://s3.amazonaws.com/dmfa-2020/project-1/hamlet.txt
```

```python
!wc  hamlet.txt
```

#### The output shows extra line not spoken by Hamlet. Hence this method does not provide the accurate output

```python
!grep -E "^HAMLET." hamlet.txt
```

```python
!grep -E "^HAMLET." hamlet.txt | wc -l
!grep -E "^POLONIUS." hamlet.txt|wc -l
!grep -E "^OPHELIA." hamlet.txt|wc -l
!grep -E "^HORATIO." hamlet.txt | wc -l
```

#### Using the below logic we were able to eliminate the extra lines and displayed only lines spoken by the mentioned characters of the play

```python
!grep "^HAMLET\." hamlet.txt
```

```python
!grep  "^HAMLET\." hamlet.txt | wc -l
!grep  "^POLONIUS\." hamlet.txt| wc -l
!grep  "^OPHELIA\." hamlet.txt| wc -l
!grep  "^HORATIO\." hamlet.txt| wc -l
```

#### The following observations were made:
* The Hamlet Play consists of 5074 lines.
* HAMLET spoke 358 times in the play
* HORATIO spoke 107 times in the play
* POLONIUS spoke 86 times in the play
* OPHELIA spoke 58 times in the play
* We conclude that Hamlet spoke maximum number of times in the play as compare to other characters mentioned here followed by Horatio.

#### Observations related to the methology of finding the above conclusion:
* There are different ways of finding the lines spoken by each character. According to me, !grep "^HAMLET\." worked best since it provided only the lines starting with uppercase followed by a period with respect to each character.
* In another logic, for example with grep -E "^HAMLET", there appears certain instances where the lines not spoken by HAMLET also appeared as shown in the text above.
